```python
# Список доступных товаров
honey_positions = [
    ("ложка", 3, 500), 
    ("солонка", 7, 1000), 
    ("свисток", 6, 200), 
    ("колокол", 1, 1000), 
    ("доска", 10, 800), 
]
def print_check(honey_positions):
    sum = 0  # переменная для накопления общей суммы
    
    # в цикле будем выводить название, количество и цену
    for honey in honey_positions:
        name = honey[0]
        amount = honey[1]
        price = honey[2]
        print(f"{name} ({amount} шт.) - {price} руб.")
        sum += amount * price  # здесь же будем считать ещё и общую сумму
    print(f"\nИтого: {sum} руб.")
    print("Спасибо за покупку!")
print_check(honey_positions)   
```

```python
from flask import Flask, render_template, request, jsonify

app = Flask(__name__)

# Список доступных товаров
honey_positions = [
    ("ложка", 3, 500),
    ("солонка", 7, 1000),
    ("свисток", 6, 200),
    ("колокол", 1, 1000),
    ("доска", 10, 800),
    ("кот в мешке", 10, 80),
    ("ендова", 10, 1500),
    ("резная ложка", 10, 1000),
    ("лопатк", 10, 500),
]

# Список для хранения купленных товаров
purchased_products = []

@app.route('/')
def index():
    return render_template('index.html')
    
@app.route('/check', methods=['POST'])
def check_product():
    data = request.json
    products_list = data.get('products', [])
    
    total_cost = 0
    not_found_products = []

    for product_name in products_list:
        found = False
        for name, quantity, price in honey_positions:
            if name == product_name:
                total_cost += price
                # Сохраняем информацию о купленном товаре
                purchased_products.append({
                    'name': name,
                    'price': price
                })
                found = True
                break
        if not found:
            not_found_products.append(product_name)

    if not_found_products:
        return jsonify({
            "status": "partial_success",
            "message": "Некоторые товары не найдены.",
            "not_found": not_found_products,
            "total_cost": total_cost
        })

    return jsonify({
        "status": "success",
        "message": f"Итого: {total_cost}. Спасибо за покупку!",
        "purchased": purchased_products  # Возвращаем список купленных товаров
    })
    
if __name__ == '__main__':
    app.run()
```
