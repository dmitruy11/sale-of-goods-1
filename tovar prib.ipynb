```python
from flask import Flask, render_template, request, jsonify

app = Flask(__name__)

# Список доступных товаров
honey_positions = [
    ("ложка", 3, 500),
    ("солонка", 7, 1000),
    ("свисток", 6, 200),
    ("колокол", 1, 1000),
    ("доска", 10, 800),
    ("кот в мешке", 10, 80),
    ("ендова", 10, 1500),
    ("резная ложка", 10, 1000),
    ("лопатка", 10, 500),
]

# Список для хранения купленных товаров
purchased_products = []

@app.route('/')
def index():
    return render_template('index.html')

@app.route('/check', methods=['POST'])
def check_product():
    data = request.json
    products_list = data.get('products', [])
    
    total_cost = 0
    not_found_products = []

    for product_name in products_list:
        found = False
        for name, quantity, price in honey_positions:
            if name == product_name:
                total_cost += price
                # Сохраняем информацию о купленном товаре
                purchased_products.append({
                    'name': name,
                    'price': price
                })
                found = True
                break
        if not found:
            not_found_products.append(product_name)

    if not_found_products:
        return jsonify({
            "status": "partial_success",
            "message": "Некоторые товары не найдены.",
            "not_found": not_found_products,
            "total_cost": total_cost
        })

    return jsonify({
        "status": "success",
        "message": f"Итого: {total_cost}. Спасибо за покупку!",
        "purchased": purchased_products  # Возвращаем список купленных товаров
    })

@app.route('/add', methods=['POST'])
def add_product():
    data = request.json
    name = data.get('name')
    quantity = data.get('quantity')
    price = data.get('price')

    if name and quantity > 0 and price > 0:
        honey_positions.append((name, quantity, price))
        return jsonify({
            "status": "success",
            "message": f"Товар '{name}' успешно добавлен."
        })

    return jsonify({
        "status": "error",
        "message": "Некорректные данные."
    })

if __name__ == '__main__':
    app.run()
```

```python
print(f"{'Название':<20} {'Цена (руб.)':<12}")  # <20 - выравнивание по левому краю, выделение 20 символов
print("=" * 32)  # Разделительная линия

# Вывод данных
for product in purchased_products:
    print(f"{product['name']:<20} {product['price']:<12.2f}")  # Вывод названия и цены с двумя знаками после запятой
```
